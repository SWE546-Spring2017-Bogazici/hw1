```python
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

# generate sample data set
N = 1000
x = np.random.randn(N)
y = -x + 0.3*np.random.randn(N)

plt.plot(x, y, '.')
plt.show()
```

```python
def plotHeatMap( dx, dy ): 
    "This function plots a heatMap representation of the sample data set for provided bindwidts dx and dy"
    heatmap, xedges, yedges = np.histogram2d(x, y, bins=(dx, dy))
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

    plt.imshow(heatmap, extent=extent, origin = 'lower')
    plt.show()
```

```python
plotHeatMap(dx=50,dy=50)
```

```python
plotHeatMap(dx=50,dy=25)
```

```python
plotHeatMap(dx=25,dy=50)
```

```python
plotHeatMap(dx=20,dy=20)
```
