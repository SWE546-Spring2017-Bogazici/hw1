```python
#Visualize dataset using a heatmap with different binwidths dx and dy
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

N = 1000
#N = 600

nBins = 50
#nBins = 5
#nBins = 100

x = np.random.randn(N)
y = np.random.randn(N)

heatmap, xx, yy = np.histogram2d(x, y, bins=nBins)


plt.imshow(heatmap,  cmap='hot', interpolation='nearest')
#plt.imshow(heatmap.T, interpolation='sinc')
plt.show()
```
